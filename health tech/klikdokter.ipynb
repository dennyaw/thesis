# SCRAPPING

```python
reviews = reviews_all(
    'id.codigo.klikdokter',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='id', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)

klikdokterraw = pd.DataFrame(np.array(reviews),columns=['review'])
klikdokterraw = klikdokterraw.join(pd.DataFrame(klikdokterraw.pop('review').tolist()))
klikdokterraw.to_csv('data_klikdokterraw.csv',index=False)
klikdokterraw
```

# DROP AND FILTER

```python
klikdokterfiltered = pd.read_csv(r'data_klikdokterraw.csv')
klikdokterfiltered['at'] = pd.to_datetime(klikdokterfiltered['at'], format='%Y-%m-%d')
#klikdokterfiltered = klikdokterfiltered.loc[(klikdokterfiltered['at'] >= "2022-1-1") & (klikdokterfiltered['at'] <= "2022-10-31")]
klikdokterfiltered = klikdokterfiltered.reset_index(drop=True)
klikdokterfiltered = klikdokterfiltered.drop(columns=['reviewId','userName','userImage','thumbsUpCount','reviewCreatedVersion','at','replyContent','repliedAt'])
klikdokterfiltered.to_csv('data_klikdokterfiltered.csv',index=False)
klikdokterfiltered
```

# CLEAN

```python
klikdoktercleaned = pd.read_csv(r'data_klikdokterfiltered.csv')

#choose reviews with 2 words or more
klikdoktercleaned = klikdoktercleaned[[len(i.split())>1 for i in klikdoktercleaned['content'].values]]

#fix contractions
klikdoktercleaned['content_contractions'] = klikdoktercleaned['content'].apply(lambda x: contractions.fix(x))

#translate the reviews to english
#klikdoktercleaned['content_translated'] = klikdoktercleaned['content_contractions'].apply(lambda x: gs.translate(x, 'en'))
klikdoktercleaned.to_csv('data_klikdoktercleaned.csv',index=False)
klikdoktercleaned
```

```python
klikdoktercleaned['lang_id'] = klikdoktercleaned['content_contractions'].apply(lambda x: detect(x))
klikdokterfinal = klikdoktercleaned.loc[klikdoktercleaned['lang_id'] == "en"]
klikdokterfinal = klikdokterfinal.drop(columns=["content",'lang_id'])
klikdokterfinal.to_csv('data_klikdokterfinal.csv',index=False)
klikdokterfinal
```

# SENTIMENT ANALYSIS

```python
klikdoktervader = pd.read_csv(r'data_klikdokterfinal.csv')
klikdoktervader['compound'] = [analyzer.polarity_scores(x)['compound'] for x in klikdoktervader['content_contractions']]
klikdoktervader['neg'] = [analyzer.polarity_scores(x)['neg'] for x in klikdoktervader['content_contractions']]
klikdoktervader['neu'] = [analyzer.polarity_scores(x)['neu'] for x in klikdoktervader['content_contractions']]
klikdoktervader['pos'] = [analyzer.polarity_scores(x)['pos'] for x in klikdoktervader['content_contractions']]


#convert rating
def getrating (compound,positive,negative):
    if compound > 0.001:
        if positive > 0.5:
            return 2
        else:
            return 1
    elif compound < -0.001:
        if negative > 0.5:
            return -2
        else:
            return -1
    elif (compound > -0.001) and (compound < 0.001):
        return 0
klikdoktervader['vader_rating'] = klikdoktervader.apply(lambda x: getrating(x.compound,x.pos,x.neg),axis=1)
klikdoktervader.to_csv('data_klikdoktervader.csv',index=False)
klikdoktervader
```
